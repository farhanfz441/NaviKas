# 🔧 FinTrack — Feature Engineering
Membuat fitur baru untuk meningkatkan performa model AI

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv('../dataset/transactions_clean.csv', parse_dates=['date'])
print(f'Data awal: {df.shape}')
```

```python
# Feature Engineering
df['is_night']      = ((df['hour'] >= 21) | (df['hour'] < 5)).astype(int)
df['is_weekend']    = df['date'].dt.dayofweek.isin([5,6]).astype(int)
df['is_high_risk']  = df['category'].isin(['shop','entertain']).astype(int)
df['log_amount']    = np.log1p(df['amount'])
df['amount_bucket'] = pd.cut(df['amount'], bins=[0,50000,200000,500000,np.inf],
                              labels=['kecil','sedang','besar','sangat_besar'])
le = LabelEncoder()
df['category_enc']  = le.fit_transform(df['category'])

print('Fitur baru berhasil dibuat:')
print(df[['amount','is_night','is_weekend','is_high_risk','log_amount','amount_bucket','is_impulsive']].head(10))
df.to_csv('../dataset/transactions_features.csv', index=False)
print(f'\nSaved: dataset/transactions_features.csv ({df.shape})')
```
